Para Merge:

union =outer

interseccion = inner

Dos dataframes los podemos tratar como un par de conjuntos, podemos encontrar un dataframe que resulte de la union o la interseccion.
Para eso uso merge

```python
estudiantes_df =  pd.DataFrame({"Nombres": ["Julia", "Miguel", "Sandra"], "Escuela":["A", "B", "C"]})
estudiantes_df
```

Unión de DataFrames

```python
pd.merge(staff_df,estudiantes_df, how="outer",left_on="Nombres", right_on="Nombres") #Une los 2 dataframes tratandolos como conjuntos uniendo toda la información posible de ambos.
```

Intersección de Dataframes

```python
pd.merge(staff_df,estudiantes_df, how="inner",left_on="Nombres", right_on="Nombres")  #Muestra sólo los nombres que se repiten en ambos conjuntos.
```

Priorización de un conjunto

Ahora nos queremos quedar con un df que sean los datos del staff y ademas si ciertas personas estudian en la uni quiero saber en que escuela

```python
pd.merge(staff_df,estudiantes_df, how="left",left_on="Nombres", right_on="Nombres") #Muestra todos los datos que hay en ambos conjuntos pero relacionados con uno solo de ellos. En este caso solo me interesa el Staff por lo que me da todos los datos del conjunto syaff y si hay algún dato de esta misma gente en el otro conjunto.
```

También se puede priorizar con how=right

Función groupby(): Agrupa elementos de un dataframe

```python
info = pd.read_csv("C:/Users/Nancy/Desktop/Phyton/Clase 3/SAmericaFruitProd.csv",  encoding = "ISO-8859-1")
info.head(2)
```

Cantidad media de frutas por cada pais en cada año.

```python
Y_C=info.groupby(["Year","Country"]) #Agrupa los paises por año.
valor_medio=Y_C["Apples","Bananas","Citrus","Kiwi","Oranges","Pears","Mandarins"].mean() #Saca el promedio de las frutas en dichos paises por año.Aquí me crea el dataframe en si.
valor_medio.head(3)
```

La cantidad total de cada fruta por pais

```python
Pais_frut=info[["Country","Apples","Bananas","Citrus","Kiwi","Oranges","Pears","Mandarins"]]
Cuadro=Pais_frut.groupby(["Country"])["Apples","Bananas","Citrus","Kiwi","Oranges","Pears","Mandarins"] #Agrupa las frutas por pais.
Cuadro.sum() #Saca la suma
```

Cantidad de Países evaluados por año

Estudiar la correlación entre las 2 mareas del problema

```python
import matplotlib.pyplot as plt
import numpy as np
#Se arma 2 series
SF=dg.SF #Es lo mismo que BA=dg["BA"]
BA=dg.BA

#Grafico ambas
plt.figure(figsize=(14,14))
plt.plot(BA)
plt.plot(SF)
```

```python
shifts=np.arange(-5,16)
plt.plot(shifts,[SF.corr(BA.shift(t)) for t in shifts])
[SF.corr(BA.shift(t)) for t in shifts]
```
